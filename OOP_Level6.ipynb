```python
class Animal:
    def eat(self):
        print("This animal eats.")

    def sleep(self):
        print("This animal sleeps.")
```

```python
class Dog(Animal):
    def bark(self):
        print("Dog barks.")
```

```python
d = Dog()

d.eat()
d.sleep()
d.bark()
```

```python
class Person:
    def __init__(self, name):
        self.name = name

    def introduce(self):
        print("My name is", self.name)


class Student(Person):
    def __init__(self, name, marks):
        super().__init__(name)   # Call parent constructor
        self.marks = marks

    def show_marks(self):
        print("Marks:", self.marks)
```

```python
s = Student("Isha", 90)

s.introduce()
s.show_marks()
```

```python
class Vehicle:
    def start(self):
        print("Vehicle started.")

    def stop(self):
        print("Vehicle stopped.")
```

```python
class Car(Vehicle):
    def honk(self):
        print("Car honks: Beep, beep!")
```

```python
my_car = Car()

print("Demonstrating Vehicle methods:")
my_car.start()
my_car.stop()

print("\nDemonstrating Car's specific method:")
my_car.honk()
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
```

```python
class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department

    def display_details(self):
        print(f"Name: {self.name}, Salary: {self.salary}, Department: {self.department}")
```

```python
# Create an instance of Manager
manager = Manager("Alice Johnson", 95000, "Sales")

# Demonstrate its attributes (properties set by the constructor)
print(f"Manager Name: {manager.name}")
print(f"Manager Salary: {manager.salary}")
print(f"Manager Department: {manager.department}")
```

```python
# Create an instance of Employee
employee = Employee("Bob Williams", 70000)

print("\n--- Employee Attributes ---")
print(f"Employee Name: {employee.name}")
print(f"Employee Salary: {employee.salary}")

print("\n--- Manager Attributes ---")
# Re-using the previously created manager instance (assuming it's still in scope)
print(f"Manager Name: {manager.name}")
print(f"Manager Salary: {manager.salary}")
print(f"Manager Department: {manager.department}")

print("\n--- Comparison ---")
print(f"Employee name matches Manager name: {employee.name == manager.name}")
print(f"Employee salary matches Manager salary: {employee.salary == manager.salary}")

# Attempting to access 'department' on Employee will raise an AttributeError,
# demonstrating that Employee does not have this attribute, unlike Manager.
try:
    print(f"Employee Department (should not exist): {employee.department}")
except AttributeError as e:
    print(f"Employee does not have a 'department' attribute: {e}")
```
