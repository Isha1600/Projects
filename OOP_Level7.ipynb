```python
class Animal:
    def sound(self):
        print("Animal sound")


class Dog(Animal):
    def sound(self):
        super().sound()
        print("Dog barks")
```

```python
d = Dog()
d.sound()
```

```python
class Person:
    def __init__(self, name):
        self.name = name

    def role(self):
        print("I am a person.")


class Student(Person):
    def role(self):
        print("I am a student.")


class Teacher(Person):
    def role(self):
        print("I am a teacher.")
```

```python
s = Student("Isha")
t = Teacher("Sir")

s.role()
t.role()
```

```python
class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def calculate_interest(self):
        return 0
```

```python
class SavingsAccount(BankAccount):
    def calculate_interest(self):
        return self.balance * 0.05
```

```python
account = BankAccount(1000)
savings = SavingsAccount(1000)

print(f"Bank Account interest: {account.calculate_interest()}")
print(f"Savings Account interest: {savings.calculate_interest()}")
```

```python
class Shape:
    def area(self):
        print("Area not defined")
```

```python
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        calculated_area = self.length * self.width
        print(f"The area of the rectangle is: {calculated_area}")
```

```python
my_rectangle = Rectangle(5, 10)
my_rectangle.area()

default_shape = Shape()
default_shape.area()
```
