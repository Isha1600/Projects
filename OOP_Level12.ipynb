```python
class Book:
    def __init__(self, pages):
        self.pages = pages

    def __add__(self, other):
        return self.pages + other.pages
```

```python
b1 = Book(100)
b2 = Book(200)

print(b1 + b2)
```

```python
class Team:
    def __init__(self, members):
        self.members = members

    def __len__(self):
        return len(self.members)
```

```python
t = Team(["A", "B", "C"])
print(len(t))
```

```python
class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def __str__(self):
        return f"Balance: ${self.balance:.2f}"

    def __add__(self, other):
        if isinstance(other, BankAccount):
            return BankAccount(self.balance + other.balance)
        return NotImplemented
```

```python
account1 = BankAccount(100.50)
account2 = BankAccount(200.75)

print(f"Account 1: {account1}")
print(f"Account 2: {account2}")

combined_account = account1 + account2
print(f"Combined Account Balance: {combined_account}")
```

```python
class MovieCollection:
    def __init__(self, movies=None):
        self.movies = movies if movies is not None else []

    def __len__(self):
        return len(self.movies)

    def __str__(self):
        if not self.movies:
            return "No movies in the collection."
        movie_titles = '\n'.join([f"- {movie}" for movie in self.movies])
        return f"Movie Collection ({len(self.movies)} movies):\n{movie_titles}"
```

```python
my_movies = MovieCollection(["The Matrix", "Inception", "Interstellar", "Dune"])

print("Length of collection:", len(my_movies))
print("\nString representation of collection:\n", my_movies)

empty_collection = MovieCollection()
print("\nEmpty collection:", empty_collection)
```
