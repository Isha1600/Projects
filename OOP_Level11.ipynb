```python
class Animal:
    def speak(self):
        print("Animal sound")

class Dog(Animal):
    def speak(self):
        print("Bark")
```

```python
class Employee:
    def work(self):
        print("Working...")

class Developer(Employee):
    def code(self):
        print("Writing code")
```

```python
class Engine:
    def start(self):
        print("Engine started")

class Car:
    def __init__(self):
        self.engine = Engine()   # Car HAS an Engine

    def start_car(self):
        self.engine.start()
        print("Car is ready")
```

```python
car = Car()
car.start_car()
```

```python
class CPU:
    def process(self):
        print("Processing data")

class Laptop:
    def __init__(self):
        self.cpu = CPU()

    def run(self):
        self.cpu.process()
        print("Laptop running")
```

```python
class Battery:
    def charge(self):
        print("Battery is charging...")

class Phone:
    def __init__(self):
        self.battery = Battery()  # Phone HAS-A Battery

    def start_phone(self):
        print("Phone is starting...")
        self.battery.charge()
```

```python
my_phone = Phone()
my_phone.start_phone()
```

```python
class Heart:
    def beat(self):
        print("Heart is beating...")

class Human:
    def __init__(self):
        self.heart = Heart()  # Human HAS-A Heart

    def feel_pulse(self):
        print("Feeling pulse...")
        self.heart.beat()
```

```python
my_human = Human()
my_human.feel_pulse()
```
